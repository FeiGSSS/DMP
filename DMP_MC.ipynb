# Twitter

```python
import pickle as pkl
with open("results/twitter_imm.pkl", "rb") as f:
    twitter_dmp_mc = pkl.load(f)

twitter_dmp = [item[0] for item in twitter_dmp_mc]
twitter_mc = [item[1] for item in twitter_dmp_mc]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
x = range(1, 51, 2)
y = range(0, 15000, 500)
plt.plot(x, twitter_mc)
plt.plot(x, twitter_dmp, marker="*")

plt.legend(["Monte Carlo (1000)", "DMP"])

plt.xlabel("#Seeds")
plt.ylabel("Influence")
plt.xticks(x)
plt.yticks(y)

plt.title("DMP vs. Monte Carlo (1000), Twitter")
plt.grid()

plt.savefig("DMP_vs_MC_Twitter.pdf")
plt.show()
```
