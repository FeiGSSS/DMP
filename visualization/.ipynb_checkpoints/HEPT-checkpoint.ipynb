```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

# Combine Overlap

```python
with open("../res/hept_imm.pkl", "rb") as f:
    imm_log = pkl.load(f)
```

```python
x = range(1, 51, 2)
plt.figure(figsize=(10, 10))
plt.plot(x, imm_log, marker="o", c="k")

plt.xticks(x)
plt.yticks(range(0, 1300, 50))
plt.legend(["IMM"])
plt.xlabel("# Seeds")
plt.ylabel("Influence")

plt.title("NetHEPT")
plt.grid(True)
plt.savefig("NetHEPT_comb_overlap.pdf")
```
