```python
import pickle as pkl
import matplotlib.pyplot as plt
```

# NetHEPT

```python
# 导入IMM的结果 {K: [seed, influence]}
with open("../SOTA_IMM/NetHEPT_imm_infs_1_51_2.pkl", "rb") as f:
    IMM_HEPT = pkl.load(f)
IMM_HEPT_XY = [list(IMM_HEPT.keys()), list([item[1] for item in IMM_HEPT.values()])]
```

```python
# 导入OPT的结果
with open("../results/NetHEPT/NetHEPT_0.0001_101_8_2_0.5_1_52_4.pkl", "rb") as f:
    OPT_HEPT = pkl.load(f)
OPT_HEPT_XY = [list(OPT_HEPT.keys()), [x[1] for x in list(OPT_HEPT.values())]]
```

```python
# 导入OPT的结果
with open("../results/NetHEPT/NetHEPT_0.0001_101_8_2_0.5_1_52_4.pkl", "rb") as f:
    TMP = pkl.load(f)
TMP_XY = [list(TMP.keys()), [x[1] for x in list(TMP.values())]]

plt.figure(figsize=(6, 5), dpi=250)

#plt
plt.plot(IMM_HEPT_XY[0], IMM_HEPT_XY[1], marker="o", c="k")
plt.plot(OPT_HEPT_XY[0], OPT_HEPT_XY[1], marker="^", c="b")
plt.plot(TMP_XY[0], TMP_XY[1], marker="^")
#prettify
## legend
legend = ["IMM", "OPT", "TMP"]
plt.legend(legend)
## title
title = "NetHEPT, N=15.2K, E=32.2K"
plt.title(title)
## ticks
plt.xticks(range(1, 52, 4))
#plt.yticks(range(0, 70000, 5000), [str(x)+"K" for x in range(0, 70, 5)])
## labels
plt.xlabel("Seed Size : K")
plt.ylabel("Influence")
## others
plt.xlim([0.5, 49.5])
plt.grid(True)
plt.savefig("./pictures/NetHEPT.png")
plt.show()
```
