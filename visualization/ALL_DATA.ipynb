```python
import pickle as pkl
import matplotlib.pyplot as plt
```

# NetHEPT

```python
# 导入IMM的结果 {K: [seed, influence]}
with open("../SOTA_IMM/NetHEPT_imm_infs_1_51_2.pkl", "rb") as f:
    IMM_HEPT = pkl.load(f)
IMM_HEPT_XY = [list(IMM_HEPT.keys()), list([item[1] for item in IMM_HEPT.values()])]
```

```python
# 导入OPT的结果
with open("../results/NetHEPT/Once_NetHEPT_0.0001_20_4_2_0.5.pkl", "rb") as f:
    OPT_HEPT = pkl.load(f)
OPT_HEPT_XY = [list(OPT_HEPT.keys()), list(OPT_HEPT.values())]
```

```python
plt.figure(figsize=(6, 5), dpi=250)

#plt
plt.plot(IMM_HEPT_XY[0], IMM_HEPT_XY[1], marker="o", c="k")
plt.plot(OPT_HEPT_XY[0], OPT_HEPT_XY[1], marker="^", c="b")
#prettify
## legend
legend = ["IMM", "OPT"]
plt.legend(legend)
## title
title = "NetHEPT, N=15.2K, E=32.2K"
plt.title(title)
## ticks
plt.xticks(range(0, 51, 5))
#plt.yticks(range(0, 70000, 5000), [str(x)+"K" for x in range(0, 70, 5)])
## labels
plt.xlabel("Seed Size : K")
plt.ylabel("Influence")
## others
plt.xlim([0.5, 49.5])
plt.grid(True)
plt.savefig("NetHEPT.png")
plt.show()
```
