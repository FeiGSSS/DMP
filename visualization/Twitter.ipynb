```python
import os
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open("../SOTA_IMM/Twitter_imm_infs.pkl", "rb") as f:
    IMM = pkl.load(f)
```

```python
IMM_INF = [x[1] for x in IMM.values()]
```

```python
plt.figure(figsize=(10,10))
x = range(1, 51, 2)
plt.plot(x, IMM_INF, marker="o", c="k")

plt.legend(["IMM","DMP_Overlap2"])

plt.xlabel("#Seeds")
plt.ylabel("Influence")
plt.xticks(x)
plt.yticks(range(0, 12000, 500))
plt.title("Twitter, n=81306, m=1768149")
plt.grid()
```

```python
Exp_Res = []
for file in os.listdir("../results/Twitter//"):
    if "influence" in file:
        Exp_Res.append(file)
```

```python
tag_data = {}
for file in Exp_Res:
    tag = file[18:93]
    with open("../results/Twitter//"+file, "rb") as f:
        data = pkl.load(f)
        tag_data[tag]=[x[1] for x in data.values()]
```

```python
plt.figure(figsize=(10, 10))

IMM_x = range(1, 51, 2)
plt.plot(IMM_x, IMM_INF, marker="o", c="k")

EXP_x = range(1, 51, 5)
for y in tag_data.values():
    if len(y) == len(range(1, 51, 2)):
        plt.plot(range(1, 51, 2), y, marker="o")
    else:
        plt.plot(EXP_x, y, marker="o")

plt.xticks(range(1, 51, 2))
plt.yticks(range(0, 13000, 500))
legend = ["IMM"]
for tag in tag_data.keys():
    legend.append(tag)
    
plt.legend(legend, bbox_to_anchor=(1.015, 1))
plt.xlabel("#Seeds")
plt.ylabel("Influence")
plt.xticks(x)
plt.yticks(range(0, 12000, 500))
plt.title("Twitter, n=81306, m=1768149")
plt.grid()
```
